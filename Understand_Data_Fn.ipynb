```python
def understand_data(df,*args, **kwargs):
    duplicate_bool = kwargs.get('duplicate_bool', None)
    import pandas as pd
    print('No of observations: '+ str(df.shape[0]))
    print('No of columns: '+ str(df.shape[1]))
    print('------------------------------------------------------------------------------------')
    print('\n')
    print('******** Describe ********* : ')
    df_des=df.describe()
    df_des=pd.DataFrame(df_des)
    print(df_des)
    print('------------------------------------------------------------------------------------')
    print('\n')
    print('******** Column data types ******** : ')
    columns=df.columns
    for col in columns:
        print( str(col) +' - '  + str(df[col].dtypes))
        if df[col].dtypes=='object':
            print('---> ' + col + ' is a categorical variable. Hence observe value_counts()')
            print(df[col].value_counts())
            print('<<<<<<>>>>>>')
    print('------------------------------------------------------------------------------------')
    print('\n')
    print('******** Null values in columns ******** : ')
    null_df=pd.isnull(df).any()
    names=['Column Names','Yes/No']
    null_df=pd.DataFrame(null_df)
    null_df.reset_index(inplace=True)
    null_df.columns=names
    print(null_df) 
    for key,value in null_df.iterrows():
        if value['Yes/No'] == True and df[value['Column Names']].dtypes!='object':
            cl_name=value['Column Names']
            cl_mean = df[cl_name].mean()
            print(cl_name + '  mean' + '  is  ' + str(cl_mean))
    print('------------------------------------------------------------------------------------')
    print('\n')
    if duplicate_bool==True:
        print('******** Duplicates in columns ******** : ')
        for col in columns:
            duplicateRowsDF = df[df.duplicated()][col]
            print("Duplicate Rows except first occurrence :")
            print(duplicateRowsDF)
        print('------------------------------------------------------------------------------------')
    print('\n')
    print('******** Head view ******** : ')
    print(df.head(3))
    print('------------------------------------------------------------------------------------')
    print('\n')
    print('******** Tail view ******** : ')
    print(df.tail(3))
    print('------------------------------------------------------------------------------------')
```

```python
def replace_nulls(df,*args, **kwargs):
    mean = kwargs.get('mean', None)
    median = kwargs.get('median', None)
    print(pd.isnull(df).any())
    columns=df.columns
    for col in columns:
        if df[col].dtypes=='object':
            print(col)
            col_mode=df[col].mode()
            print(col_mode)
            df[col]=df[col].fillna(col_mode)[0]
        if df[col].dtypes!='object':
            if mean==True:
                print(col)
                cl_mean = df[col].mean()
                print(cl_mean)
                df[col]=df[col].fillna(cl_mean)
            if median==True:
                print(col)
                col_median=df[col].median()
                print(cl_mean)
                df[col]=df[col].fillna(col_median)
            else:
                print(col)
                cl_mean = df[col].mean()
                print(cl_mean)
                df[col]=df[col].fillna(cl_mean)
                
    print('Null values check')
    print(pd.isnull(df).any())
    return df
```
